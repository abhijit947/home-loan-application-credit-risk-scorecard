### 3A. Bureau Tradeline Data

Independent features will be created from bureau tradeline data for all loans, segmented by on-us and off-us accounts, by loan type, and by on-us/off-us * loan-type combinations.

```python
import numpy as np
import pandas as pd
```

```python
bureau_tradeline_df['Remaining_Months']=bureau_tradeline_df['Tenure_Months']-bureau_tradeline_df['Account_Age_Months']
```

#### 3Ai. All Loans

```python
bureau_features=pd.DataFrame(bureau_tradeline_df['Customer_ID'].drop_duplicates().reset_index(drop=True))

loan_cnt = (bureau_tradeline_df
            .groupby("Customer_ID")
            .size()
            .reset_index(name="All_Loan_Count"))

bureau_features = bureau_features.merge(loan_cnt, on="Customer_ID", how="left")

agg_func = ["sum", "mean","median", "max", "min"]
cust = bureau_tradeline_df.groupby("Customer_ID")[["Original_Loan_Amount","Balance"]].agg(agg_func)
cust=cust.reset_index()
cust.columns = [f"All_{col}_{func}" for col, func in cust.columns]
cust=cust.rename(columns={"All_Customer_ID_":"Customer_ID"})
bureau_features=bureau_features.merge(cust,on='Customer_ID',how='left')


agg_func = ["max", "min"]
cust = bureau_tradeline_df.groupby("Customer_ID")[["Tenure_Months","Account_Age_Months","Remaining_Months"]].agg(agg_func)
cust=cust.reset_index()
cust.columns = [f"All_{col}_{func}" for col, func in cust.columns]
cust=cust.rename(columns={"All_Customer_ID_":"Customer_ID"})
bureau_features=bureau_features.merge(cust,on='Customer_ID',how='left')

agg_func = ["sum","max"]
cust = bureau_tradeline_df.groupby("Customer_ID")[["DPD30p_L3M",
                                  "DPD60p_L3M",
                                  "DPD90p_L3M",
                                  "DPD30p_L6M",
                                  "DPD60p_L6M",
                                  "DPD90p_L6M",
                                  "DPD30p_L12M",
                                  "DPD60p_L12M",
                                  "DPD90p_L12M"]].agg(agg_func)
cust=cust.reset_index()
cust.columns = [f"All_{col}_{func}" for col, func in cust.columns]
cust=cust.rename(columns={"All_Customer_ID_":"Customer_ID"})
bureau_features=bureau_features.merge(cust,on='Customer_ID',how='left')
```

#### 3Aii. By on-us and off-us accounts

```python
bureau_features1=pd.DataFrame(bureau_tradeline_df['Customer_ID'].drop_duplicates().reset_index(drop=True))
Lender_Flag_US_vs_OFFUS_unique=bureau_tradeline_df['Lender_Flag_US_vs_OFFUS'].unique()

for i in Lender_Flag_US_vs_OFFUS_unique:

    df=bureau_tradeline_df[bureau_tradeline_df['Lender_Flag_US_vs_OFFUS']==i].copy()

    loan_cnt = (df.groupby("Customer_ID")
                  .size()
                  .reset_index(name=f"{i}_Loan_Count"))
    bureau_features1 = bureau_features1.merge(loan_cnt, on="Customer_ID", how="left")
    
    agg_func = ["sum", "mean","median", "max", "min"]
    cust1 = df.groupby("Customer_ID")[["Original_Loan_Amount","Balance"]].agg(agg_func)
    cust1.columns = [f"{i}_{col}_{func}" for col, func in cust1.columns]
    cust1=cust1.reset_index()

    bureau_features1=bureau_features1.merge(cust1,on='Customer_ID',how='left')
    
    
    agg_func = ["max", "min"]
    cust1 = df.groupby("Customer_ID")[["Tenure_Months","Account_Age_Months","Remaining_Months"]].agg(agg_func)
    cust1.columns = [f"{i}_{col}_{func}" for col, func in cust1.columns]
    cust1=cust1.reset_index()
    bureau_features1=bureau_features1.merge(cust1,on='Customer_ID',how='left')
    
    agg_func = ["sum","max"]
    cust1 = df.groupby("Customer_ID")[["DPD30p_L3M",
                                      "DPD60p_L3M",
                                      "DPD90p_L3M",
                                      "DPD30p_L6M",
                                      "DPD60p_L6M",
                                      "DPD90p_L6M",
                                      "DPD30p_L12M",
                                      "DPD60p_L12M",
                                      "DPD90p_L12M"]].agg(agg_func)
    
    cust1.columns = [f"{i}_{col}_{func}" for col, func in cust1.columns]
    cust1=cust1.reset_index()
    bureau_features1=bureau_features1.merge(cust1,on='Customer_ID',how='left')
```

#### 3Aiii. By loan type

```python
bureau_features2=pd.DataFrame(bureau_tradeline_df['Customer_ID'].drop_duplicates().reset_index(drop=True))
Tradeline_Type_unique=bureau_tradeline_df['Tradeline_Type'].unique()

for i in Tradeline_Type_unique:

    df=bureau_tradeline_df[bureau_tradeline_df['Tradeline_Type']==i].copy()

    loan_cnt = (df.groupby("Customer_ID")
                .size()
                .reset_index(name=f"{i}_Loan_Count"))
    
    bureau_features2 = bureau_features2.merge(loan_cnt, on="Customer_ID", how="left")
    
    agg_func = ["sum", "mean","median", "max", "min"]
    cust2 = df.groupby("Customer_ID")[["Original_Loan_Amount","Balance"]].agg(agg_func)
    cust2.columns = [f"{i}_{col}_{func}" for col, func in cust2.columns]
    cust2=cust2.reset_index()

    bureau_features2=bureau_features2.merge(cust2,on='Customer_ID',how='left')
    
    
    agg_func = ["max", "min"]
    cust2 = df.groupby("Customer_ID")[["Tenure_Months","Account_Age_Months","Remaining_Months"]].agg(agg_func)
    cust2.columns = [f"{i}_{col}_{func}" for col, func in cust2.columns]
    cust2=cust2.reset_index()
    bureau_features2=bureau_features2.merge(cust2,on='Customer_ID',how='left')
    
    agg_func = ["sum","max"]
    cust2 = df.groupby("Customer_ID")[["DPD30p_L3M",
                                      "DPD60p_L3M",
                                      "DPD90p_L3M",
                                      "DPD30p_L6M",
                                      "DPD60p_L6M",
                                      "DPD90p_L6M",
                                      "DPD30p_L12M",
                                      "DPD60p_L12M",
                                      "DPD90p_L12M"]].agg(agg_func)
    
    cust2.columns = [f"{i}_{col}_{func}" for col, func in cust2.columns]
    cust2=cust2.reset_index()
    bureau_features2=bureau_features2.merge(cust2,on='Customer_ID',how='left')
```

#### 3Aiv By on-us/off-us * loan-type combinations

```python
bureau_features3=pd.DataFrame(bureau_tradeline_df['Customer_ID'].drop_duplicates().reset_index(drop=True))

for i in Lender_Flag_US_vs_OFFUS_unique:
    
    for j in Tradeline_Type_unique:

        df=bureau_tradeline_df[(bureau_tradeline_df['Lender_Flag_US_vs_OFFUS']==i) & (bureau_tradeline_df['Tradeline_Type']==j)].copy()

        loan_cnt = (df.groupby("Customer_ID")
                      .size()
                      .reset_index(name=f"{i}_{j}_Loan_Count"))
        bureau_features3 = bureau_features3.merge(loan_cnt, on="Customer_ID", how="left")
        
        agg_func = ["sum", "mean","median", "max", "min"]
        cust3 = df.groupby("Customer_ID")[["Original_Loan_Amount","Balance"]].agg(agg_func)
        cust3.columns = [f"{i}_{j}_{col}_{func}" for col, func in cust3.columns]
        cust3=cust3.reset_index()
    
        bureau_features3=bureau_features3.merge(cust3,on='Customer_ID',how='left')
        
        
        agg_func = ["max", "min"]
        cust3 = df.groupby("Customer_ID")[["Tenure_Months","Account_Age_Months","Remaining_Months"]].agg(agg_func)
        cust3.columns = [f"{i}_{j}_{col}_{func}" for col, func in cust3.columns]
        cust3=cust3.reset_index()
        bureau_features3=bureau_features3.merge(cust3,on='Customer_ID',how='left')
        
        agg_func = ["sum","max"]
        cust3 = df.groupby("Customer_ID")[["DPD30p_L3M",
                                          "DPD60p_L3M",
                                          "DPD90p_L3M",
                                          "DPD30p_L6M",
                                          "DPD60p_L6M",
                                          "DPD90p_L6M",
                                          "DPD30p_L12M",
                                          "DPD60p_L12M",
                                          "DPD90p_L12M"]].agg(agg_func)
        
        cust3.columns = [f"{i}_{j}_{col}_{func}" for col, func in cust3.columns]
        cust3=cust3.reset_index()
        bureau_features3=bureau_features3.merge(cust3,on='Customer_ID',how='left')
```
